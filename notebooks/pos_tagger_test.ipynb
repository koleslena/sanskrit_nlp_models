```python
import torch
from sanskrit_tagger.tagger_factory import get_pos_tagger

device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = torch.hub.load('koleslena/sanskrit_nlp_models', 'pos_tagger_model', force_reload=True)

tagger = get_pos_tagger(model, device=device)

test_splitter_sentences = ["agnihotrāt samutthāya harṣeṇa mahatā anvitā",
                            "uvāca ca enam vara-dā vacanam pārthivam tadā"]

for sent, predicted in zip(test_splitter_sentences, tagger(test_splitter_sentences)):
    print(f"Input:  {sent}")
    print(f"Output: {predicted}")
```
