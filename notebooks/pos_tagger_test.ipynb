```python
import torch
from sanskrit_tagger.tagger_factory import get_pos_tagger

device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = torch.hub.load('koleslena/sanskrit_nlp_models', 'pos_tagger_model', version="model-20260502-2325", force_reload=True)

tagger = get_pos_tagger(model, device=device)

test_pos_sentences = ["agnihotrāt samutthāya harṣeṇa mahatā anvitā",
                        "uvāca ca enam vara-dā vacanam pārthivam tadā"]

test_pos_sentences_tokenized = [sent.split() for sent in test_pos_sentences]

for sent_tokens, sent_tags in zip(test_pos_sentences_tokenized, tagger(test_pos_sentences)):
    print(' '.join('{}-{}'.format(tok, tag) for tok, tag in zip(sent_tokens, sent_tags)))
    print()
```

```python
from sanskrit_tagger.tagger_factory import get_segmenter

model = torch.hub.load('koleslena/sanskrit_nlp_models', 'segmenter_model', force_reload=True)

segmenter = get_segmenter(model, device=device)

test_splitter_sentences = ["agnihotrātsamutthāya harṣeṇa mahatānvitā",
                            "uvāca cainaṁ varadā vacanaṁ pārthivaṁ tadā"]

right = """
        agnihotrāt samutthāya harṣeṇa mahatā anvitā |
        uvāca ca enam vara-dā vacanam pārthivam tadā 
        """

pred = []

for sent, predicted in zip(test_splitter_sentences, segmenter(test_splitter_sentences)):
    print(f"Input:  {sent}")
    print(f"Output: {predicted}")
    pred.append(predicted)


pos_model = torch.hub.load('koleslena/sanskrit_nlp_models', 'pos_tagger_model', version="model-20260502-2325", force_reload=True)

tagger = get_pos_tagger(pos_model, device=device)

pred_cleaned = [p.replace("-", " ") for p in pred]

test_pos_sentences_tokenized = [sent.split() for sent in pred_cleaned]

for sent_tokens, sent_tags in zip(test_pos_sentences_tokenized, tagger(test_pos_sentences)):
    print(' '.join('{}-{}'.format(tok, tag) for tok, tag in zip(sent_tokens, sent_tags)))
    print()
```
