```python
import torch
from sanskrit_tagger.tagger_factory import get_segmenter

device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = torch.hub.load('koleslena/sanskrit_nlp_models:f64f54b', 'segmenter_model', version='model-20260501-2131', device=device, force_reload=True)

segmenter = get_segmenter(model, device=device)

test_splitter_sentences = ["agnihotrātsamutthāya harṣeṇa mahatānvitā",
                            "uvāca cainaṁ varadā vacanaṁ pārthivaṁ tadā"]

right = """
        agnihotrāt samutthāya harṣeṇa mahatā anvitā |
        uvāca ca enam vara-dā vacanam pārthivam tadā 
        """

for sent, predicted in zip(test_splitter_sentences, segmenter(test_splitter_sentences)):
    print(f"Input:  {sent}")
    print(f"Output: {predicted}")
```
